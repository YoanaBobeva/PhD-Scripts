```python
print(merged_df.columns)
```

```python
# Define columns you want to keep
columns_to_keep = ['Gene names', 't_stat', 'p_value', 'Mean Control', 'Mean ALS',
                   'p_adjusted_BH', 'p_adjusted_bonf', 'FC ALS-C', 'Log2FC ALS-C',
                   'Uniprot', 'Protein family', 'SS', 'Regulation']
```

```python
# Filter each DataFrame and collect into a list
filtered_dfs = [df[columns_to_keep] for df in df_list]  # Make sure to keep it as a DataFrame

# Merge them into one
merged = pd.concat(filtered_dfs, ignore_index=True, axis=1)

# Save the 'Gene Name' column to an Excel file
merged.to_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\top_80_abs_all datasets.xlsx", index=False)
```

```python
# Standardize each DataFrame to have all desired columns, filling missing ones with NaN
standardized_dfs = []
for df in df_list:
    temp = df.copy()
    for col in columns_to_keep:
        if col not in temp.columns:
            temp[col] = pd.NA  # Add missing column as NA
    temp = temp[columns_to_keep]  # Reorder/trim columns
    standardized_dfs.append(temp)

# Concatenate all into one DataFrame
merged_df = pd.concat(standardized_dfs, ignore_index=True)
```

```python
# Save the 'Gene Name' column to an Excel file
merged_df.to_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\top_80_abs_all datasets.xlsx", index=False)
```
