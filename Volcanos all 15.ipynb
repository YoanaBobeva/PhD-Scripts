```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
c1=pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD Shared folder\Datasets\Set C1\CSF extra volcano.xlsx")
c1.head()
```

```python
#Define regulation
def regulation_category(row):
    if row['p_value'] < 0.05 and row['Log2FC'] > 0:
        return 'Upregulated'
    elif row['p_value'] < 0.05 and row['Log2FC'] < 0:
        return 'Downregulated'
    else:
        return 'Not significant'



```

```python
#Add the regulation column
c1['Regulation'] = c1.apply(regulation_category, axis=1)

# Add -log10(p-value)
c1['-log10(p-value)'] = -np.log10(c1['p_value'])

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=c1,
    x='Log2FC',
    y='-log10(p-value)',
    hue='Regulation',
    palette={'Upregulated': 'red', 'Downregulated': 'blue', 'Not significant': 'gray'},
    legend='full'
)

# Threshold lines
plt.axhline(-np.log10(0.05), linestyle='--', color='black')
plt.axvline(0, linestyle='--', color='black')

# Label significant proteins
for _, row in c1.iterrows():
    if row['Regulation'] != 'Not significant':
        plt.text(row['Log2FC'], row['-log10(p-value)'], row['Gene names'],
                 fontsize=8, ha='right', va='bottom')

plt.title('Volcano plot: Set C1')
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10 p-value')
plt.legend(title='Regulation')
plt.tight_layout()
plt.show()
```
