```python
import pandas as pd
from gseapy import enrichr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csm=pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Enrichment in-silico.xlsx", sheet_name="Top 80 CSM", header=0, index_col="Gene names")
csm.head()
```

```python
ss=pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Enrichment in-silico.xlsx", sheet_name="Top 80 SS", header=0, index_col="Gene names")
ss.head()
```

```python
both=pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Enrichment in-silico.xlsx", sheet_name="Both top 80", header=0, index_col="Gene names")
both.head()
```

```python
# Convert the gene_name column to list
gene_list = csm.index.tolist()

gsea_results = enrichr(gene_list=gene_list,
                       gene_sets='GO_Biological_Process_2021',
                       organism='Human',
                       outdir='gsea_output')
# Load results
res_df = gsea_results.res2d
top_hits = res_df.sort_values('Adjusted P-value', ascending=True).head(15)

# Plot the figure
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=top_hits,
    x='Combined Score',
    y='Term',
    size='Overlap',  # You can also parse this to get the number of genes
    hue='Adjusted P-value',
    palette='coolwarm_r',
    sizes=(50, 300),
    legend='brief'
)
plt.title('Enrichment analyses of the top 80 CSM')
plt.xlabel('Combined Score')
plt.ylabel('GO Term')
plt.grid(True)

# Move legend outside the figure
plt.legend(
    bbox_to_anchor=(1.05, 1),   # X=1.05 pushes it outside to the right
    loc='upper left',
    borderaxespad=0
)

plt.tight_layout()
plt.show()
```
