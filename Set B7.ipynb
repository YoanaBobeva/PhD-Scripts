```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from numpy import sqrt
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests
```

```python
char= pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD Shared folder\Datasets\Set B7\Char ALC HC.xlsx", header=0, index_col="Gene names")
char
```

```python
# Automatically identify group columns by their prefixes
group1= char.loc[:, char.columns.str.startswith('Control ')]
group2= char.loc[:, char.columns.str.startswith('ALS ')]
```

```python
# Flatten all values in the DataFrame to a single array
all_values = char.values.flatten()

# Remove NaNs if any
all_values = all_values[~np.isnan(all_values)]

# Calculate min, max, and range
min_val = np.min(all_values)
max_val = np.max(all_values)
value_range = max_val - min_val

print("Minimum value:", min_val)
print("Maximum value:", max_val)
print("Range:", value_range)

```

```python
# Perform Mann-Whitney U-test row-wise, drop NA values 
#stat, p_value = mannwhitneyu(group1.iloc[0], group2.iloc[0], alternative='two-sided', method='auto')

stats = []
p_vals = []
for i in range(group1.shape[0]):
    stat, p_value = mannwhitneyu(group1.iloc[i].dropna(), group2.iloc[i].dropna(), alternative='two-sided', method='auto')
    stats.append(stat)
    p_vals.append(p_value)
```

```python
char['U-stat'] = stats
char['P-value'] = p_vals
char.head()
```

```python
# Apply Benjamini-Hochberg correction (optional)
_, p_adjusted, _, _ = multipletests(char['P-value'], method='fdr_bh')
char['p_adjusted_BH'] = p_adjusted
char.head()
```

```python
_, p_adjusted, _, _ = multipletests(char['P-value'], method='bonferroni')
char['p_adjusted_bonf'] = p_adjusted
char.head()
```

```python
char['modified_BH'] = char['p_adjusted_BH'].apply(lambda x: 0.0001 if x < 0.0001 else x)
```

```python
char['modified_bonf'] = char['p_adjusted_bonf'].apply(lambda x: 0.0001 if x < 0.0001 else x)
char.head()
```

```python
#df['average'] = df.mean(numeric_only=True, axis=1)
char['Mean Control'] = char.loc[:, char.columns.str.startswith('Control ')].mean(axis=1)
char['Mean ALS'] = char.loc[:, char.columns.str.startswith('ALS ')].mean(axis=1)
```

```python
#Fold Change
char["FC ALS-C"]= char["Mean ALS"]/char["Mean Control"]
```

```python
#Log 2 FC
char["Log2FC ALS-C"]=np.log2(char["FC ALS-C"])
char.head()
```
