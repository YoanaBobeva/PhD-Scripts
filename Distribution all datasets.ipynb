```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from numpy import sqrt
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
c1= pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD Shared folder\Datasets\Set C1\CSF extracellular original.xlsx", 
                    header=0, index_col= "Gene names")
c1.head()
```

```python
c1.plot.hist(bins=30, figsize=(15, 10), edgecolor="black")
plt.title("Set C1 data distribution", fontsize=16)
plt.xlabel('Values', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
c1.stack().plot.hist(bins=60, figsize=(15, 10), edgecolor="black")
plt.title("Set C1 data distribution", fontsize=16)
plt.xlabel('Normalised protein abundance values', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
stat, p_value = shapiro(c1)

if p_value > 0.05:
    print("Data is normally distributed (fail to reject H0)")
else:
    print("Data is not normally distributed (reject H0)")
```
