```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from numpy import sqrt
from statsmodels.stats.multitest import multipletests
from scipy.stats import ttest_ind
```

```python
disc= pd.read_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD Shared folder\Datasets\Set C12\Discovery original no stats.xlsx", header=0, index_col = 'Gene names'
                      )
disc
```

```python
# Reverse log2 transformation (calculate 2^Log p-value)
prot = 2 ** disc
prot.head()
```

```python
prot.shape
```

```python
# Columns belonging to Group 1 and 2
ALS_columns = [col for col in prot.columns if "ALS" in col]
Control_columns = [col for col in prot.columns if "Control" in col]


# Filter rows where at least 50% of Group 1 values are present, do the same for Group 2
filtered_als = prot[prot[ALS_columns].notna().mean(axis=1) >= 0.5]
cont_filt = filtered_als[filtered_als[Control_columns].notna().mean(axis=1) >= 0.5]
```

```python
cont_filt.shape
```

```python
cont_filt.to_excel(r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Discovery of biomarkers\Discovery distribution check.xlsx")
```

```python
# Automatically identify group columns by their prefixes
als = [col for col in cont_filt.columns if col.startswith('ALS_')]
control = [col for col in cont_filt.columns if col.startswith('Control_')]

# Perform t-tests row-wise
results = []
for i, row in cont_filt.iterrows():
    t_stat, p_value = ttest_ind(row[als].dropna(), row[control].dropna())
    results.append({'Gene names': i, 't_stat': t_stat, 'p_value': p_value})
    
# Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df
```

```python
# Add the results DataFrame to the original data DataFrame
prot_stat = pd.concat([cont_filt, results_df.set_index('Gene names')], axis=1)
prot_stat.head()
```

```python
# Apply Benjamini-Hochberg correction (optional)
_, p_adjusted, _, _ = multipletests(prot_stat['p_value'], method='fdr_bh')
prot_stat['p_adjusted_bh'] = p_adjusted
prot_stat.head()
```

```python
# Adjust p-values using the Bonferroni method
_, p_adjusted, _, _ = multipletests(prot_stat['p_value'], method='bonferroni')

# Add the adjusted p-values to the DataFrame
prot_stat['p_adjusted_bon'] = p_adjusted
```

```python
#df['average'] = df.mean(numeric_only=True, axis=1)
prot_stat['Mean Control'] = prot_stat.loc[:, prot_stat.columns.str.startswith('Control_')].mean(axis=1)
prot_stat['Mean ALS'] = prot_stat.loc[:, prot_stat.columns.str.startswith('ALS_')].mean(axis=1)
```

```python
#Fold Change
prot_stat["FC ALS-C"]= prot_stat["Mean ALS"]/prot_stat["Mean Control"]
```

```python
#Log 2 FC
prot_stat["Log2FC ALS-C"]=np.log2(prot_stat["FC ALS-C"])
```
