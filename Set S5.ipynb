```python
# Replace 0 with NaN
olf.replace(0, np.nan, inplace=True)
```

```python
# Columns belonging to Group 1 and 2
ALS_columns = [col for col in olf.columns if "ALS" in col]
Control_columns = [col for col in olf.columns if "Control" in col]


# Filter rows where at least 50% of Group 1 values are present, do the same for Group 2
filtered_als = olf[olf[ALS_columns].notna().mean(axis=1) >= 0.5]
cont_filt = filtered_als[filtered_als[Control_columns].notna().mean(axis=1) >= 0.5]
```

```python
#cont_filt = pd.read_excel (r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Olfactory\Olfactory drop NA rows.xlsx", 
              header=0, index_col = "Gene names")
#cont_filt.head()
```

```python
# Set the new range
min_range = 1
max_range = 5
```

```python
dis_min = cont_filt.min().min()
dis_max = cont_filt.max().max()
scaled_dis = (cont_filt - dis_min) / (dis_max - dis_min)
scaled_data = (scaled_dis * (max_range - min_range)) + min_range
scaled_data.head()
```

```python
scaled_data.to_excel (r"C:\Users\Client\OneDrive - Queen Mary, University of London\PhD\PhD back up\Programing full\R\Datasets\Olfactory\Olfactory drop NA scaled.xlsx")
```

```python
# Automatically identify group columns by their prefixes
als = [col for col in scaled_data.columns if col.startswith('ALS ')]
control = [col for col in scaled_data.columns if col.startswith('Control ')]

# Perform t-tests row-wise
results = []
for i, row in scaled_data.iterrows():
    t_stat, p_value = ttest_ind(row[als].dropna(), row[control].dropna())
    results.append({'Gene names': i, 't_stat': t_stat, 'p_value': p_value})
    
# Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df
```

```python
# Add the results DataFrame to the original data DataFrame
prot_stat = pd.concat([scaled_data, results_df.set_index('Gene names')], axis=1)
prot_stat.head()
```

```python
# Apply Benjamini-Hochberg correction (optional)
_, p_adjusted, _, _ = multipletests(prot_stat['p_value'], method='fdr_bh')
prot_stat['p_adjusted_bh'] = p_adjusted
prot_stat.head()
```

```python
_, p_adjusted, _, _ = multipletests(prot_stat['p_value'], method='bonferroni')
prot_stat['p_adjusted_bonf'] = p_adjusted
prot_stat.head()
```

```python
#df['average'] = df.mean(numeric_only=True, axis=1)
prot_stat['Mean Control'] = prot_stat.loc[:, prot_stat.columns.str.startswith('Control ')].mean(axis=1)
prot_stat['Mean ALS'] = prot_stat.loc[:, prot_stat.columns.str.startswith('ALS ')].mean(axis=1)
```

```python
#Fold Change
prot_stat["FC ALS-C"]= prot_stat["Mean ALS"]/prot_stat["Mean Control"]
```

```python
#Log 2 FC
prot_stat["Log2FC ALS-C"]=np.log2(prot_stat["FC ALS-C"])
prot_stat.head()
```
